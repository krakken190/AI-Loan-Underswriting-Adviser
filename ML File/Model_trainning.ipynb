```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
import pickle

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("synthetic_credit_data (1).csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Here created an instance of LabelEncoder
label_encoder = LabelEncoder()

df['VoterRegistration'] = label_encoder.fit_transform(df['VoterRegistration'])

df['PropertyOwnership'] = label_encoder.fit_transform(df['PropertyOwnership'])
```

```python
# Defining features and target
features = ['Salary', 'Assets', 'ExistingLoans', 'CreditScore', 'NumAccounts', 
            'PropertyOwnership', 'VoterRegistration', 'TotalIncome', 'Fraud','Creditworthy']
target_credit = 'Creditworthy'
target_fraud = 'Fraud'

X = df[features]
y_credit = df[target_credit]
y_fraud = df[target_fraud]

X_train_credit, X_test_credit, y_train_credit, y_test_credit = train_test_split(X, y_credit, test_size=0.2, random_state=42)
X_train_fraud, X_test_fraud, y_train_fraud, y_test_fraud = train_test_split(X, y_fraud, test_size=0.2, random_state=42)

# Model trainnning
credit_model = RandomForestClassifier()
credit_model.fit(X_train_credit, y_train_credit)

fraud_model = RandomForestClassifier()
fraud_model.fit(X_train_fraud, y_train_fraud)

with open('credit_worthiness_model.pkl', 'wb') as f:
    pickle.dump(credit_model, f)

with open('fraud_detect_ion_model.pkl', 'wb') as f:
    pickle.dump(fraud_model, f)
```
